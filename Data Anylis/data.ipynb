```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")

# Create the figure with 2 rows and 2 columns of plots
fig, axes = plt.subplots(2, 2, figsize=(8, 10))

# 1. Survival rate by Passenger Class
sns.barplot(x='Pclass', y='Survived', data=df, estimator=lambda x: sum(x)/len(x), ax=axes[0,0])
axes[0,0].set_title('Survival Rate by Passenger Class')
axes[0,0].set_ylabel('Survival Rate')
axes[0,0].set_xlabel('Passenger Class (1 = Highest)')

# 2. Survival rate by Gender
sns.barplot(x='Sex', y='Survived', data=df, estimator=lambda x: sum(x)/len(x), ax=axes[0,1])
axes[0,1].set_title('Survival Rate by Gender')
axes[0,1].set_ylabel('Survival Rate')
axes[0,1].set_xlabel('Gender')

# 3. Survival rate by Family Size
sns.barplot(x='FamilySize', y='Survived', data=df, estimator=lambda x: sum(x)/len(x), ax=axes[1,0])
axes[1,0].set_title('Survival Rate by Family Size')
axes[1,0].set_ylabel('Survival Rate')
axes[1,0].set_xlabel('Family Size')

# 4. Survival rate by IsAlone
sns.barplot(x='isAlone', y='Survived', data=df, estimator=lambda x: sum(x)/len(x), ax=axes[1,1])
axes[1,1].set_title('Survival Rate - Alone vs With Family')
axes[1,1].set_ylabel('Survival Rate')
axes[1,1].set_xlabel('Is Alone (1 = Alone, 0 = With Family)')

# Adjust layout for better spacing
plt.tight_layout()
plt.show()
```

```python
def age_group(Age):
    if Age < 18:
        return 'child'
    elif Age < 60:
        return 'Adult'
    else:
        return 'Senior'
```

```python
df['AgeGroup']=df['Age'].apply(age_group) 
```

Passenger Class: Survival rates were highest in 1st class, followed by 2nd class, and lowest in 3rd class. This suggests socio-economic status played a major role in survival chances.

Gender: Female passengers had a significantly higher survival rate than males, reflecting the "women and children first" rescue policy.

Family Size: Survival rates varied by family size. Small families (2-4 members) tended to have better survival rates compared to passengers alone or in very large families. This could indicate support from family had a positive effect.

Traveling Alone (IsAlone): Passengers traveling alone had a noticeably lower survival rate compared to those with family aboard, reinforcing the importance of social support during the disaster.
